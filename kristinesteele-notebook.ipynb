## Section 4. Create a Seaborn Scatter Plot

A scatter plot is used to visualize the relationship between two numerical variables. 
This is a Markdown cell. 

````python
# This cell is Python too - previous cells must be run first

# Create a scatter plot of penguin flipper length vs. body mass
# We need to know our dataset column names so we can use them to set x, y, and hue (color)
scatter_plt: Axes = sns.scatterplot(
    data=penguins_df, x="flipper_length_mm", y="body_mass_g", hue="species"
)

# Set axis labels using the Matplotlib Axes methods set_xlabel() and set_ylabel()
scatter_plt.set_xlabel("Flipper Length (mm)")
scatter_plt.set_ylabel("Body Mass (g)")

# Set the title using the Matplotlib Axes set_title() method
scatter_plt.set_title("Chart 1. Penguin Flipper Length vs. Body Mass (by Species)")

# Run this cell by clicking the play button (Execute cell and below) in the cell toolbar
# Or by pressing Shift+Enter
# Or by clicking Run ALL in the Menu above
````

````python
import matplotlib.pyplot as plt
import numpy as np

def plot_sine(freq):
    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(freq * x)
    plt.plot(x, y)
    plt.title(f"Sine Wave (freq={freq})")
    plt.show()

widgets.interact(plot_sine, freq=widgets.IntSlider(min=1, max=10, step=1, value=1))
````

### Section 10. Advanced Widget Integration: Diamonds 

*some comment about a created graph

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_df = sns.load_dataset("diamonds")

# Show basic statistics
print(diamonds_df.describe())

# Example plot: price vs carat, colored by cut
sns.scatterplot(data=diamonds_df, x="carat", y="price", hue="cut", alpha=0.5)
plt.title("Diamond Price vs Carat by Cut")
plt.show()
````

````python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm, Normalize

# Interactive hexbin plot for diamonds dataset
def interactive_hexbin(gridsize=60, cmap='viridis', log_scale=True):
    plt.figure(figsize=(10,8))
    norm = LogNorm() if log_scale else Normalize()
    hb = plt.hexbin(diamonds_df['carat'], diamonds_df['price'], gridsize=gridsize, cmap=cmap, mincnt=1, norm=norm)
    plt.xlabel('Carat')
    plt.ylabel('Price')
    scale_text = 'Log Color Scale' if log_scale else 'Linear Color Scale'
    plt.title(f'Diamond Price vs Carat (Hexbin Plot, {scale_text})')
    cb = plt.colorbar(hb)
    cb.set_label(f'Count in bin ({"log" if log_scale else "linear"} scale)')
    plt.tight_layout()
    plt.show()

gridsize_slider = widgets.IntSlider(value=60, min=20, max=100, step=5, description='Grid Size')
cmap_dropdown = widgets.Dropdown(options=['viridis', 'plasma', 'Blues', 'inferno', 'magma'], value='viridis', description='Colormap')
logscale_toggle = widgets.Checkbox(value=True, description='Log Scale')

widgets.interact(interactive_hexbin, gridsize=gridsize_slider, cmap=cmap_dropdown, log_scale=logscale_toggle)
````

### Section 10. Advanced Widget Integration: Diamonds

*Interactively explore diamond prices by cut using the dropdown below.*

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets

# Load the diamonds dataset
diamonds_df = sns.load_dataset("diamonds")

# Dropdown for cut selection
cut_options = diamonds_df['cut'].dropna().unique().tolist()
cut_dropdown = widgets.Dropdown(
    options=cut_options,
    value=cut_options[0],
    description='Cut:',
    style={'description_width': 'initial'}
)

def plot_diamonds_by_cut(selected_cut):
    filtered = diamonds_df[diamonds_df['cut'] == selected_cut]
    plt.figure(figsize=(7,5))
    plt.scatter(filtered['carat'], filtered['price'], c='purple', alpha=0.5)
    plt.xlabel('Carat')
    plt.ylabel('Price')
    plt.title(f'Diamond Price vs Carat for {selected_cut} Cut')
    plt.grid(True)
    plt.show()

widgets.interact(plot_diamonds_by_cut, selected_cut=cut_dropdown)
